## DATA CLEANING

```python
Movies = Movies[(Movies['Income'] != 'Unknown') & (Movies['Budget'] != 'Unknown')]
```

```python
for idx, value in enumerate(Movies['Income']):
  if str(Movies.iloc[idx].Income)[0] == '$':
    Movies.loc[Movies['Income'] == value, 'Income'] = int(sub("[^0-9]", "", value))
```

```python
conversion_dict = {
    '$': 1.0,
    '€': 1.07,
    '£': 1.21,
    '₩': 0.00078,
    'C': 0.73,
    '₹': 0.012,
    '¥': 0.0075,
    'A': 0.68,
    'S': 0.095,
    'D': 0.14,
    'N': 0.10 
}
```

```python
for idx, value in enumerate(Movies['Budget']):
  if str(Movies.iloc[idx].Budget)[0] in conversion_dict.keys():
    Movies.loc[Movies['Budget'] == value, 'Budget'] = int(sub("[^0-9]", "", value)) * conversion_dict[Movies.iloc[idx].Budget[0]]
```

```python
Movies
```

```python
# Calculate the profit column by subtracting the Budget from the Income
Movies['Profit'] = Movies['Income'] - Movies['Budget']
```

```python
# Sort the dataframe by Profit column in descending order and select the top 10 rows
top_profit_movies = Movies.sort_values('Profit', ascending=False).head(10)
```

```python
# Plot a bar chart of the top 10 movies by profit
sns.barplot(data=top_profit_movies, x='Profit', y='Title')
plt.title('Top 10 movies by profit')
plt.xlabel('Profit (in millions of dollars)')
plt.ylabel('Movie title')
plt.show()
sns.set(rc={'figure.figsize':(30,12)})
```

# 0.5

Genre analysis: You can group the movies by genre and analyze the distribution of ratings, budget, and income within each genre. This can help you understand which genres are most popular with audiences and critics, and how different genres perform financially.

```python
# Group the dataframe by Genre and calculate the mean Rating, Budget, and Income for each genre
genre_stats = Movies.groupby('Genre').agg({'Rating': 'mean', 'Budget': 'mean', 'Income': 'mean'}).reset_index()
```

```python
genre_stats
```

```python
# Plot a bar chart of the mean Rating, Budget, and Income for each genre
sns.set(style='whitegrid')
sns.barplot(data=genre_stats, x='Genre', y='Rating', color='blue')
sns.barplot(data=genre_stats, x='Genre', y='Budget', color='red')
sns.barplot(data=genre_stats, x='Genre', y='Income', color='green')
plt.title('Average ratings, budget, and income by genre')
plt.xlabel('Genre')
plt.ylabel('Amount (in millions of dollars or rating points)')
plt.show()
sns.set(rc={'figure.figsize':(30,12)})
```

Director analysis: You can group the movies by director and analyze the distribution of ratings, budget, and income for each director. This can help you identify the most successful and critically acclaimed directors in the dataset.
