```python
#imports

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

#regex
import re
```

```python
#importing the clean csv data file
df=pd.read_csv('BA_Data_Collection.csv')
```

# cleaning ✅ Trip Verified | from Reviews

```python
#removing ✅ Trip Verified | 
remove = "✅ Trip Verified |"
df["reviews"] = df["reviews"].str.replace(remove, "")
```

```python
df
```

```python
#this sign "|" is still visible in the review section
remove = "|"
df["reviews"] = df["reviews"].str.replace(remove, "")
```

```python
#Cleared
df
```

# Cleaning Ratings with Stars

```python
#check for unique values
df.stars.unique()
```

```python
# remove the \t and \n from the ratings
df.stars = df.stars.str.strip("\n\t\t\t\t\t\t\t\t\t\t\t\t\t")
```

#### there are 3 rows that are None. we will drop all these 3 rows

```python
drop= "None"

df=df[df['stars'] != drop]
```
